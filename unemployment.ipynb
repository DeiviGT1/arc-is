```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

csv = pd.read_csv('excel/Unemployemnt_series_statewise_USA.csv')
```

```python
def order(csv):
  csv = pd.read_csv('excel/Unemployemnt_series_statewise_USA.csv')
  csv.columns.values[0] = "Fecha"
  csv.set_index(csv["Fecha"], inplace=True)
  csv = pd.melt(csv, id_vars=['Fecha'], var_name='Estado', value_name='Valor')
  csv['Fecha'] = pd.to_datetime(csv['Fecha'])
  return csv
```

```python
csv = pd.read_csv('excel/Unemployemnt_series_statewise_USA.csv')
csv = order(csv)

# csv = csv[csv["Estado"].str.contains("Texas")]

csv.plot(
    x="Fecha", y="Valor", figsize=(15, 5)
    )

#get the value where its max
max = csv[csv["Valor"] == csv["Valor"].max()]
plt.axvline(x=max["Fecha"].values[0], color='r', linestyle='--', linewidth=2)
#annotate the max value
plt.annotate(
    np.datetime_as_string(max["Fecha"].values[0], unit='D'), 
    xy=(max["Fecha"].values[0], max["Valor"].values[0]),
    xytext=(max["Fecha"].values[0], max["Valor"].values[0] + 1),
    arrowprops=dict(facecolor='black', shrink=0.05),
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"),
    )
plt.show()
```
